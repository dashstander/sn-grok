```python
from sngrok.permutations import Permutation
from sngrok.groups import generate_subgroup
from functools import reduce


def union(cosets):
    if len(cosets) == 0:
        return cosets
    else:
        return reduce(lambda x, y: set(x).union(set(y)), cosets)


s3_x_s2 = {
    "order": 12,
    "copies": 10,
    "generators": [
        [(0, 1, 3, 4, 2), (0, 1, 3, 2, 4), (1, 0, 2, 3, 4)], #1
        [(0, 3, 2, 4, 1), (0, 1, 2, 4, 3), (2, 1, 0, 3, 4)], #2
        [(0, 2, 4, 3, 1), (0, 2, 1, 3, 4), (3, 1, 2, 0, 4)], #3
        [(0, 2, 3, 1, 4), (0, 2, 1, 3, 4), (4, 1, 2, 3, 0)], #4
        [(3, 1, 2, 4, 0), (3, 1, 2, 0, 4), (0, 2, 1, 3, 4)], #5
        [(2, 1, 4, 3, 0), (2, 1, 0, 3, 4), (0, 3, 2, 1, 4)], #6
        [(2, 1, 3, 0, 4), (2, 1, 0, 3, 4), (0, 4, 2, 3, 1)], #7
        [(1, 4, 2, 3, 0), (1, 0, 2, 3, 4), (0, 1, 3, 2, 4)], #8
        [(1, 3, 2, 0, 4), (1, 0, 2, 3, 4), (0, 1, 4, 3, 2)], #9
        [(1, 2, 0, 3, 4), (1, 0, 2, 3, 4), (0, 1, 2, 4, 3)], #10
    ]

}

frobenius_groups = {
    "order": 20,
    "copies": 6,
    "generators": [
        [(0, 2, 3, 4, 1), (1, 2, 4, 0, 3)],
        [(0, 2, 3, 4, 1), (1, 4, 3, 0, 2)],
        [(0, 2, 4, 1, 3), (1, 2, 3, 4, 0)],
        [(0, 2, 4, 1, 3), (1, 3, 4, 2, 0)],
        [(0, 3, 4, 2, 1), (1, 3, 0, 4, 2)],
        [(0, 3, 4, 2, 1), (1, 4, 0, 2, 3)]
    ]
    
}


s4_subgroups = {
    "order": 24,
    "copies": 5,
    "generators": [
        [(0, 2, 3, 4, 1), (0, 2, 1, 3, 4)], #1 fix 0
        [(2, 1, 3, 4, 0), (2, 1, 0, 3, 4)], #2 fix 1
        [(1, 3, 2, 4, 0), (1, 0, 2, 3, 4)], #3 fix 2
        [(1, 2, 4, 3, 0), (1, 0, 2, 3, 4)], #4 fix 3
        [(1, 2, 3, 0, 4), (1, 0, 2, 3, 4)], #5 fix 4
    ]
}


s3_subgroups = {
    "order": 6,
    "copies": 10,
    "generators": [
        [(0, 1, 3, 4, 2), (0, 1, 3, 2, 4)], #1  fix 0 1
        [(0, 3, 2, 4, 1), (0, 1, 2, 4, 3)], #2  fix 0 2
        [(0, 2, 4, 3, 1), (0, 2, 1, 3, 4)], #3  fix 0 3
        [(0, 2, 3, 1, 4), (0, 2, 1, 3, 4)], #4  fix 0 4
        [(3, 1, 2, 4, 0), (3, 1, 2, 0, 4)], #5  fix 1 2
        [(2, 1, 4, 3, 0), (2, 1, 0, 3, 4)], #6  fix 1 3
        [(2, 1, 3, 0, 4), (2, 1, 0, 3, 4)], #7  fix 1 4
        [(1, 4, 2, 3, 0), (1, 0, 2, 3, 4)], #8  fix 2 3
        [(1, 3, 2, 0, 4), (1, 0, 2, 3, 4)], #9  fix 2 4
        [(1, 2, 0, 3, 4), (1, 0, 2, 3, 4)], #10 fix 3 4
    ]
}

a4_subgroups = {
    "order": 12,
    "copies": 5,
    "generators": [
        [(0, 2, 1, 4, 3), (0, 3, 1, 2, 4)], # 1 fix 0
        [(2, 1, 0, 4, 3), (0, 1, 4, 2, 3)], # 2 fix 1
        [(1, 0, 2, 4, 3), (3, 0, 2, 1, 4)], # 3 fix 2
        [(1, 0, 4, 3, 2), (2, 0, 1, 3, 4)], # 4 fix 3
        [(1, 0, 3, 2, 4), (2, 0, 1, 3, 4)], # 5 fix 4

    ]
}


z5_groups = {
    "order": 5,
    "copies": 6,
    "generators": [
        [(1, 2, 4, 0, 3)],
        [(1, 4, 3, 0, 2)],
        [(1, 2, 3, 4, 0)],
        [(1, 3, 4, 2, 0)],
        [(1, 3, 0, 4, 2)],
        [(1, 4, 0, 2, 3)]
    ]
}


z6_groups = {
    "order": 6,
    "copies": 10,
    "generators": [
        [(0, 1, 3, 4, 2), (1, 0, 2, 3, 4)], #1  fix 0 1
        [(0, 3, 2, 4, 1), (2, 1, 0, 3, 4)], #2  fix 0 2
        [(0, 2, 4, 3, 1), (3, 1, 2, 0, 4)], #3  fix 0 3
        [(0, 2, 3, 1, 4), (4, 1, 2, 3, 0)], #4  fix 0 4
        [(3, 1, 2, 4, 0), (0, 2, 1, 3, 4)], #5  fix 1 2
        [(2, 1, 4, 3, 0), (0, 3, 2, 1, 4)], #6  fix 1 3
        [(2, 1, 3, 0, 4), (0, 4, 2, 3, 1)], #7  fix 1 4
        [(1, 4, 2, 3, 0), (0, 1, 3, 2, 4)], #8  fix 2 3
        [(1, 3, 2, 0, 4), (0, 1, 4, 3, 2)], #9  fix 2 4
        [(1, 2, 0, 3, 4), (0, 1, 2, 4, 3)], #10 fix 3 4
    ]
}


def get_left_cosets(subgroup, full_n):
    Sn =  Permutation.full_group(full_n)
    all_sn = set([s.sigma for s in Sn])
    perms = [s for s in subgroup]
    cosets = set()
    while union(cosets) != all_sn:
        left = Sn.pop()
        coset = tuple(sorted([(left * s).sigma for s in perms]))
        cosets.add(coset)
    return cosets



def get_right_cosets(subgroup, full_n):
    Sn =  Permutation.full_group(full_n)
    all_sn = set([s.sigma for s in Sn])
    perms = [s for s in subgroup]
    cosets = set()
    while union(cosets) != all_sn:
        right = Sn.pop()
        coset = tuple(sorted([(s * right).sigma for s in perms]))
        cosets.add(coset)
    return cosets



```

```python
class PerfectDoubleCosetCircuit:
    def __init__(self, n: int, lsubgroup: list[tuple[int]], rsubgroup: list[tuple[int]]):
        self.left_subgroup = [Permutation(p) for p in lsubgroup]
        self.right_subgroup = [Permutation(p) for p in rsubgroup]
        
        self.left_cosets = get_left_cosets(self.right_subgroup, n)
        self.right_cosets = get_right_cosets(self.left_subgroup, n)
        
        self.index = len(self.left_cosets)
        
        assert self.index == len(self.right_cosets)
        
        double_coset = self.left_cosets.intersection(self.right_cosets)
                
        self.double_coset = list(double_coset)[0]
        
    
    def get_coset(self, perm: Permutation):
        for coset in self.left_cosets:
            if perm.sigma in coset:
                return coset

    def left_perm_coset_membership(self, perm: Permutation) -> Permutation:
        """
        Returns a coset representative of the right coset of `left_subgroup`
        `perm` is in.
        """
        for coset in self.right_cosets:
            if perm.sigma in coset:
                return Permutation(coset[0])

    def right_perm_coset_membership(self, perm: Permutation) -> Permutation:
        """
        Returns a coset representative of the left coset of `right_subgroup`
        `perm` is in.
        """
        for coset in self.left_cosets:
            if perm.sigma in coset:
                return Permutation(coset[0])

    def coset_multiplication(self, left_perm, right_perm):
        left_coset_rep = self.left_perm_coset_membership(left_perm)
        right_coset_rep = self.right_perm_coset_membership(right_perm)
        if (left_coset_rep * right_coset_rep).sigma in self.double_coset:
            return 0.
        else:
            return 1.
```

```python
from copy import deepcopy
import random


class DoubleCosetCircuit:

    def __init__(self, n: int, lsubgroup: list[tuple[int]], rsubgroup: list[tuple[int]]):
        self.left_subgroup = [Permutation(p) for p in lsubgroup]
        self.right_subgroup = [Permutation(p) for p in rsubgroup]
        
        self.left_cosets = get_left_cosets(self.right_subgroup, n)
        self.right_cosets = get_right_cosets(self.left_subgroup, n)
        
        self.index = len(self.left_cosets)
        
        assert self.index == len(self.right_cosets)
        
        double_coset = self.left_cosets.intersection(self.right_cosets)
                
        self.double_coset = list(double_coset)[0]
        self.right_perm_vals, self.left_perm_vals = self._get_coset_vals()
    
    def _get_coset_vals(self):
        left_vals = {}
        right_vals = {}
        
        double_coset_rep = Permutation(self.double_coset[0])
        right_cosets = [deepcopy(coset) for coset in self.right_cosets]
        
        for coset in self.left_cosets:
            value = random.uniform(-1., 1.)
            left_rep = Permutation(coset[0])
            left_vals[left_rep.sigma] = value
            
            right_rep = (double_coset_rep * left_rep.inverse).sigma
            for i, right_coset in enumerate(right_cosets):
                if right_rep in right_coset:
                    right_vals[right_coset[0]] = -1. * value
                    right_cosets.pop(i)
                    break
        return left_vals, right_vals

    
    def left_perm_coset_membership(self, perm: Permutation) -> Permutation:
        """
        Returns a coset representative of the right coset of `left_subgroup`
        `perm` is in.
        """
        for coset in self.right_cosets:
            if perm.sigma in coset:
                return coset[0]

    def right_perm_coset_membership(self, perm: Permutation) -> Permutation:
        """
        Returns a coset representative of the left coset of `right_subgroup`
        `perm` is in.
        """
        for coset in self.left_cosets:
            if perm.sigma in coset:
                return coset[0]
    
    def inverse_circuit(self):
        circuit = deepcopy(self)
        circuit.left_perm_vals = {k: -1.0 * v for k, v in self.left_perm_vals.items()}
        circuit.right_perm_vals = {k: -1.0 * v for k, v in self.right_perm_vals.items()}
        return circuit
            
    def coset_multiplication(self, left_perm, right_perm):
        left_coset_rep = self.left_perm_coset_membership(left_perm)
        right_coset_rep = self.right_perm_coset_membership(right_perm)
        return max(self.left_perm_vals[left_coset_rep] + self.right_perm_vals[right_coset_rep], 0.0)
```

```python
import math

def circuit_trial(n, lsubgroup, rsubgroup, num_circuits):
    Sn = Permutation.full_group(n)
    group_order = 1.0 * len(Sn)
    perfect_circuit = PerfectDoubleCosetCircuit(n, lsubgroup, rsubgroup)
    circuits = [DoubleCosetCircuit(n, lsubgroup, rsubgroup) for _ in range(num_circuits)]
    num_misclassified = 0
    num_properly_classified = 0
    mse = 0.0
    for lperm, rperm in product(Sn, Sn):
        perf_val = perfect_circuit.coset_multiplication(lperm, rperm)
        circuit_val = min(sum([c.coset_multiplication(lperm, rperm) for c in circuits]), 1.0)
        if circuit_val == 0 and perf_val == 1:
            num_misclassified += 1
        
        if perf_val == 0 and abs(circuit_val) < 0.00001:
            num_properly_classified += 1
            
        mse += (perf_val - circuit_val) ** 2
    return {
        'index': str(circuits[0].index),
        'num_circuits': num_circuits,
        'num_misclassified': num_misclassified,
        'num_correct': num_properly_classified,
        'false_negative_ratio': num_misclassified / (num_properly_classified + num_misclassified),
        'pcnt_misclassified': num_misclassified / (group_order**2),
        'total_error': mse,
        'mean_sqare_error': mse / (group_order**2)
    }
    
    
    
```
